```python
import pandas as pd

path = '/content/drive/MyDrive/northstar_dataset/'

customers = pd.read_csv(path + 'customers.csv')
orders = pd.read_csv(path + 'orders.csv')
deliveries = pd.read_csv(path + 'deliveries.csv')
complaints = pd.read_csv(path + 'complaints.csv')
incidents = pd.read_csv(path + 'incidents.csv')

print("Datasets loaded")
```

```python
customers_collection.insert_many(
    customers.to_dict("records")
)

print("Customers inserted")
```

```python
orders_collection.insert_many(
    orders.to_dict("records")
)

print("Orders inserted")
```

```python
deliveries_collection.insert_many(
    deliveries.to_dict("records")
)

print("Deliveries inserted")
```

```python
complaints_collection.insert_many(
    complaints.to_dict("records")
)

print("Complaints inserted")
```

```python
incidents_collection.insert_many(
    incidents.to_dict("records")
)

print("Incidents inserted")
```

```python
new_document = {
    "complaint_id": "CP1001",
    "customer_id": "C100",
    "complaint_type": "Late Delivery",
    "severity": "High"
}

complaints_collection.insert_one(new_document)

print("Document inserted")
```

```python
complaints_collection.update_one(
    {"complaint_id": "CP0001"},
    {"$set": {"severity": "Resolved"}}
)

print("Document updated")
```

```python
complaints_collection.delete_one(
    {"complaint_id": "CP1001"}
)

print("Document deleted")
```

```python
deliveries_collection.create_index("delivery_status")

deliveries_collection.create_index("driver_id")

complaints_collection.create_index("customer_id")

incidents_collection.create_index("delivery_id")

print("Indexes created")
```

```python
explain_output = deliveries_collection.find(
    {"delivery_status": "Failed"}
).explain()

print(explain_output)
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$delivery_status",
            "total": {"$sum": 1}
        }
    }
]

results = deliveries_collection.aggregate(pipeline)

for r in results:
    print(r)
```
